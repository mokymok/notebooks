**The following is a script for a histogram showing the correlation of tip percentages (based on subtotal of the bill) and sex.**

```python
#define a new dataframe with the sex and tip_percentage columns  
df_stp = pd.DataFrame(
    {
        'sex_stp':df['sex'],
        'tip_percentage_stp':df['tip_percentage']

    }
)

#define two more DataFrames that separate sexes prior to recombining in a
#third new DataFrame that we'll then use for plotting the histogram
df_stp.sort_values('sex_stp')
df_stp_f = df_stp.loc[df_stp['sex_stp']=='Female']
df_stp_m = df_stp.loc[df_stp['sex_stp']=='Male']

#modify first new df (df_stp_f) to have 1-step int-value indexes
df_stp_f.index = pd.RangeIndex(start=0, stop=87, step=1)
df_stp_f.rename(index=int).index
pd.Index(
    [0, 1, 2], 
    dtype='object'
)

#modify second new df (df_stp_m) to have 1-step int-value indexes
df_stp_m.index = pd.RangeIndex(start=0, stop=156, step=1)
df_stp_m.rename(index=int).index
pd.Index(
    [0, 1, 2], 
    dtype='object'
)

#define the final DataFrame that will be used to plot the histogram
df_tp = pd.DataFrame(
    {
        'mt':df_stp_m['tip_percentage_stp'],
        'ft':df_stp_f['tip_percentage_stp']
    }
)

#plot the histogram  
df_tp.plot.hist(
    title='tip percentage among men and women',
    bins=70,
    alpha=.75
)
```

**A slightly modified version is below, evening out the amount of data from both men and women, since there are about twice as many entries for men as women.**

```python
df_tp_mod = df_tp.dropna()
df_tp_mod.plot.hist(    
    title='modified tip percentage among men and women',
    bins=100,
    alpha=.75
)
```

**The following script shows the correlation between time of day (lunch and dinner) and the total amount presented on the bill.**

```python
#define a new dataframe with the time and total_bill columns  
df_ttb = pd.DataFrame(
    {
        'time_ttb':df['time'],
        'total_bill_ttb':df['total_bill']

    }
)

#define two more DataFrames that separate times prior to recombining in a
#third new DataFrame that we'll then use for plotting the scatter plot
df_ttb.sort_values('time_ttb')
df_ttb_d = df_ttb.loc[df_ttb['time_ttb']=='Dinner']
df_ttb_l = df_ttb.loc[df_ttb['time_ttb']=='Lunch']

#modify first new df (df_ttb_d) to have 1-step int-value indexes
df_ttb_d.index = pd.RangeIndex(start=0, stop=175, step=1)
df_ttb_d.rename(index=int).index
pd.Index(
    [0, 1, 2], 
    dtype='object'
)

#modify second new df (df_ttb_l) to have 1-step int-value indexes
df_ttb_l.index = pd.RangeIndex(start=0, stop=68, step=1)
df_ttb_l.rename(index=int).index
pd.Index(
    [0, 1, 2], 
    dtype='object'
)

#define the final DataFrame that will be used to plot the histogram
df_ttbf = pd.DataFrame(
    {
        'dinner':df_ttb_d['total_bill_ttb'],
        'lunch':df_ttb_l['total_bill_ttb']
    }
)

#plot the histogram  
df_ttbf.plot.hist(
    title='total bill based on meal time',
    bins=70,
    alpha=.59
)
```

**Altair, another library, is great for visualizing geometries, both 2d and 3d, and is also built on the same Pandas DataFrames I've been using.**

**Below, though, is a couple examples of what you can do with it (not using the current dataset.)**  

**The first is graticules (the latitudinal and longitudinal lines you see on a globe) from an orthographic projection.**

```python
data_grat = alt.graticule(step=[15, 15])

alt.Chart(data_grat).mark_geoshape(stroke='black').project(
    'orthographic',
    rotate=[0, -45, 0]
)
```

**Likewise, we can turn this "sphere" any which way just by editing the "rotate" args.**

**We can also change the "step" args to increase or decrease the distance between the lines.**

The projection style can be one of the following:
'albers', 
'albersUsa', 
'azimuthalEqualArea', 
'azimuthalEquidistant', 
'conicConformal', 
'conicEqualArea', 
'conicEquidistant', 
'equalEarth', 
'equirectangular', 
'gnomonic', 
'identity', 
'mercator', 
'naturalEarth1', 
'orthographic', 
'stereographic', 
'transverseMercator'.

```python
data_1 = alt.graticule(step=[47, 6])

alt.Chart(data_1).mark_geoshape(stroke='blue').project(
    'equalEarth',
    rotate=[14, 125, 72]
)
```

```python
alt.Chart(data_1).mark_geoshape(stroke='blue').project(
    'orthographic',
    rotate=[14, 125, 72]
)
```

**Although Altair is great for visualizing geometries, its main purpose is actually to streamline conversion of plot specifications to a JSON string that complies with the Vega-Lite schema.  In order to see exactly what Altair is sending to Vega-Lite, we can use the "to_json()" method.**

```python
print(chart_dfj.to_json())
```

**Although this is its main purpose, it's also great for making interactive charts and graphs that allow you to see your data in different ways.**

```python
brush = alt.selection_interval()

chart_int = alt.Chart(df).mark_point().encode(
    y='total_bill:Q',
    color=alt.condition(brush, 'sex:N', alt.value('lightgray'))
).properties(
    width=600,
    height=600
).add_selection(
    brush
)

chart_int.encode(x='tip:Q') | chart_int.encode(x='tip_percentage:Q')
```
