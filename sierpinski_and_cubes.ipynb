```python
n=16
def a(n):
 y=-~n
 while y>=0:
  x=i=0
  while(i<y):
   print(" ",end="");i=i+1
  while(x+y<n):
   if((x&y)!=0):
    print(" ",end=" ")
   else:print("▲ ",end="")
   x=x+1
  print()
  y=y-1
a(n)
```
