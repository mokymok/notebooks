the rules of the Van eck sequence are as follows:

    the first term is zero.
    If the last term is new then the next term is zero.
    otherwise the next term is how far back the term occured previously.

the first few terms are as follows:

    0         (using a)
    0 0       (using b)
    0 0 1     (using c)
    0 0 1 0   (using b)
    0 0 1 0 2 (using c)

Print the first 1000 terms of the Van eck sequence, with each term on a new line.

```python
A=[0];l={}
for i in range(1000):
    n=i-l.get(A[i],i)
    A.append(n)
    l[A[i]]=i
print(A)
```
